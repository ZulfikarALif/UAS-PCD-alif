```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#baca gambar
jeruk=cv2.imread('jeruk.jpg')
#convert BGR ke RGB
jeruk=cv2.cvtColor(jeruk,cv2.COLOR_BGR2RGB)
#convert RGK ke HSV
hsv_jeruk=cv2.cvtColor(jeruk, cv2.COLOR_RGB2HSV)
```

```python
#deklarasi batas bawah (orange cerah)
light_orange=(1,190,200)
#deklarasi batas atas (light dark)
dark_orange =(18, 255, 255)
```

```python
#tresholding
mask=cv2.inRange(hsv_jeruk,light_orange,dark_orange)
#impose gambar asli dengan mask
result=cv2.bitwise_and(jeruk,jeruk, mask=mask)
```

```python
#plotiing
plt.subplot(1,4,1)
plt.imshow(jeruk)
plt.subplot(1,4,2)
plt.imshow(hsv_jeruk)
plt.subplot(1,4,3)
plt.imshow(mask)
plt.subplot(1,4,4)
plt.imshow(result)
plt.show()
```
